````python
import pandas as pd
import numpy as np

# creating the data set
data = {
    'Species': ['Homo sapiens', 'Mus musculus', 'Drosophila melanogaster', 'Homo sapiens', 'Mus musculus', 'Drosophila melanogaster'],
    'Gene': ['GeneA', 'GeneA', 'GeneA', 'GeneB', 'GeneB', 'GeneB'],
    'Expression_Level': [5.3, 7.1, 6.5, 4.8, 5.9, 6.0],
    'Sample_Size': [100, 120, 110, 95, 115, 105],
    'Origin': ['USA (100)', 'Germany (120)', 'Japan (110)', 'USA (95)', 'Germany (115)', 'Japan (105)']
}

df = pd.DataFrame(data)

print(df)
print()


def bootstrap_population_mean(df, n_iterations=1000, random_seed=42): # This function takes in the data set, the number of bootstrap iterations, and the random seed to ensure this can be reproduced
    np.random.seed(random_seed) # This ensures we always use the same seed to make the same random numbers
    means = [] # This is the mean values from each bootstrapped sample 
    
    expression_data = df['Expression_Level'] # This takes the expression_levels column from the data frame. This is what we resample when we perform bootstrapping

    # Bootstrapping
    for _ in range(n_iterations): # This takes n_iterations number of samples from the expression_data column
        sample = np.random.choice(expression_data, size=len(expression_data), replace=True) # This samples from the data WITH REPLACEMENT (to make it independent), the size = len(expression_data) resamples the same number of data points as in the original data set and replace = True keeps replacing the data
        means.append(np.mean(sample)) # This calculates the mean

    original_mean = np.mean(expression_data) # This calculates the original mean (without bootstrapping)
    ci_lower = np.percentile(means, 2.5) # This calculates the lower bound of the 95% confidence interval
    ci_upper = np.percentile(means, 97.5) # This calculates the upper bound
    
    return original_mean, ci_lower, ci_upper

original_mean, ci_lower, ci_upper = bootstrap_population_mean(df) 
print(f"Original Mean: {original_mean}")
print(f"95% Confidence Interval: [{ci_lower}, {ci_upper}]")
````

Background: Firstly, the difference between the population parameter and the sample statistic is that the population parameter is the true value of some characteristic (e.g. the population mean); however, it is usually unknown because the population is too large to be able to measure the characteristic directly. So instead, we use the sample statistic which estimates the population parameter and is from a smaller subsample of the population. The sample statistic, because it is taken from random samples of the population, can vary which is known as sampling variability. Thus there is an "error" in the estimation by the sample compared to the true parameter. So, confidence intervals are used to indicate how confident we are that the population parameter lies within a range around the sample statistic.

Direct answer: The distinction matters because the sample statistic changes but the population parameter is fixed. Because the sample statistic can fluctuate, that is why we have confidence intervals as the interval represents the range in which we think the population parameter lies. It represents the uncertainty of the sample statistic. This means that we cannot say "there is a 95% chance that the population parameter lies in the confidence interval" because the parameter is either in or out of the interval. This insinuates that the population parameter moves but it is the sample statistic that varies. Instead we should say that "if we did the sample many times, 95% of the confidence intervals for each sample would contain the true population parameter".

### 1. **Distinguishing Population Parameters from Sample Statistics**:
   - **Population Parameter**: A fixed, unknown value that describes a characteristic of an entire population (e.g., population mean \(\mu\)).
   - **Sample Statistic**: A value calculated from a sample that estimates the population parameter (e.g., sample mean \(\bar{x}\)).
   - This distinction is important for confidence intervals because confidence intervals use the sample statistic to estimate the population parameter, accounting for the uncertainty introduced by sampling variability.

---

### 2. **Importance of Distinction in Confidence Intervals**:
   - **Sample Statistic**: Varies from sample to sample due to randomness and is used to estimate the population parameter.
   - **Population Parameter**: A fixed, unknown value we aim to estimate with confidence intervals.
   - Confidence intervals provide a range within which the true population parameter is likely to lie, based on the sample statistic. This range accounts for sampling variability.

---

### 3. **Sample Variability**:
   - Sample variability refers to the natural variation in sample statistics (e.g., sample mean or sample proportion) when different samples are drawn from the same population.
   - Larger samples tend to reduce sample variability, leading to more precise estimates of the population parameter.
   - **Standard Error (SE)**: A measure of sample variability, reflecting how much the sample statistic is expected to vary across different samples.

---

### 4. **Why "95% Chance" is Incorrect in Confidence Intervals**:
   - It is incorrect to say, "There is a 95% chance that the population parameter lies in the confidence interval."
   - Once the confidence interval is calculated from a sample, the population parameter either is or is not within the interval.
   - The correct interpretation of a 95% confidence interval is: "If we repeatedly sampled from the population and constructed confidence intervals, 95% of those intervals would contain the true population parameter."
   - The 95% refers to the long-run frequency of confidence intervals containing the true parameter, not the probability that any one particular interval contains it.

1. **Why do we fail to reject the null hypothesis when the confidence interval covers 0?**
   - When a confidence interval includes 0, it suggests that 0 is a plausible value for the true population parameter (e.g., mean difference or effect size). This means there is insufficient evidence to reject the null hypothesis, which typically states that there is no effect (i.e., the parameter is 0).

2. **Why does the null hypothesis mean the population mean is 0?**
   - The null hypothesis doesn't always mean the population mean is 0. Instead, it asserts no effect or difference in many common tests. For example, in a two-sample t-test, the null hypothesis might state that the difference between two population means is 0, meaning no significant difference exists between the groups.

3. **Why does \( \mu = 0 \) imply no effect?**
   - In hypothesis testing, when the population mean (or effect size) is hypothesized as 0, it implies no effect or no difference between groups. For instance, in a test comparing two groups, if \( \mu_1 - \mu_2 = 0 \), it suggests no difference between the group means, meaning no effect of the treatment or condition being tested.

4. **Why do we fail to reject the null hypothesis when the confidence interval covers 0, even if the sample mean is not 0?**
   - The confidence interval accounts for the uncertainty in estimating the true population mean. Even if the sample mean is not 0, if the confidence interval includes 0, it indicates that 0 could still be the true population mean due to variability in the sample data. Therefore, we fail to reject the null hypothesis because we can't confidently rule out that the true mean is 0.

#### Data Visualization 2 : 

This histogram how many patients felt a difference in health scores. It allows us to see which patients felt better or worse after taking the vaccine and by how much (e.g. if the health score difference is 5, the patient felt significantly better after using the drug).

````python
import pandas as pd
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42)

# Sample data
data = pd.read_csv("https://raw.githubusercontent.com/sylvia-wong2/STA130_HW4/refs/heads/main/patient_health_data.csv")

# Create DataFrame
df = pd.DataFrame(data)

# Calculate the difference between final and initial health scores
df['HealthScoreDifference'] = df['FinalHealthScore'] - df['InitialHealthScore']

# Bootstrapping function
def bootstrap(data, n_iterations=10000):
    boot_means = []
    n_size = len(data)
    
    for _ in range(n_iterations):
        # Sample with replacement
        sample = np.random.choice(data, size=n_size, replace=True)
        # Calculate the mean of the bootstrapped sample
        boot_means.append(np.mean(sample))
    
    return boot_means

# Bootstrapping the health score differences
bootstrapped_means = bootstrap(df['HealthScoreDifference'].values)

# Display the bootstrapped means
bootstrapped_means[:10]  # Display the first 10 bootstrapped means for inspection
````

#### Quantitative Analysis (methodology code and explanation):

In the block of code above, I have used bootstrapping to sample from the data 10000 times. I used bootstrapping because from this bootstrapped data, I can calculate the confidence interval to determine the uncertainty of my sample statistics. From this confidence interval, I can make an inference about the population mean. That is, I expect that if I did the sample many times, I am confident that 95% of the confidence intervals will contain the population mean.

````python
# Calculate the 95% confidence interval
confidence_level = 0.95
alpha = 1 - confidence_level
lower_bound = np.percentile(bootstrapped_means, 100 * (alpha / 2))
upper_bound = np.percentile(bootstrapped_means, 100 * (1 - alpha / 2))

# Create a histogram with Plotly
fig = px.histogram(
    x=bootstrapped_means,
    nbins=30,
    title='Histogram of Bootstrapped Sample Means',
    labels={'x': 'Bootstrapped Mean Health Score Difference'},
)

# Calculate mean of bootstrapped means
mean_bootstrapped = np.mean(bootstrapped_means)

# Add a vertical line for the mean of the bootstrapped means with value annotation
fig.add_vline(
    x=mean_bootstrapped, 
    line_dash='dash', 
    line_color='red',
    annotation_text=f'Mean: {mean_bootstrapped:.2f}', 
    annotation_position='top right'
)


# Add vertical lines for the confidence interval bounds
fig.add_vline(x=lower_bound, line_color='blue', line_dash='dash', annotation_text=f'Lower Bound: {lower_bound}', annotation_position='top right')
fig.add_vline(x=upper_bound, line_color='blue', line_dash='dash', annotation_text=f'Upper Bound: {upper_bound}', annotation_position='top right')
# Show the figure
fig.show(renderer = "png")
````

#### Quantitative Analysis (Visualization)

In this histogram, we can see that the bootstrapped data that we had are clustered around a mean of 3.31 which means that in general, after we have taken around 10000 samples from the population, most of our samples have an average of 3.31. In other words, on average, people said that they felt slightly better after the vaccine. Only a little number of people said they felt a lot better or not at all better. 

Between the two blue lines represents the 95% confidence interval from 0.9 to 5.6. This means that we are 95% confident that we can find the population mean in between 0.9 and 5.6.

Thus mu cannot take the value of 0 which is enough to reject the null hypothesis. This means that the vaccine DOES have an affect on the people. It does not have no effect.
